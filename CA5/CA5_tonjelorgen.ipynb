### Imports

````python
# Importing necessary libraries and packages
import pandas as pd

# Importing method for plotting and visualizing data
import seaborn as sns
import matplotlib.pyplot as plt

# Hiding warnings
import warnings
warnings.filterwarnings('ignore')
````

### Reading data

````python
# Loading the data

chili_train = pd.read_csv('./data/train.csv', delimiter=',')

chili_test = pd.read_csv('./data/test.csv', delimiter=',')
````

````python
chili_train.head(5)
````

````python
chili_test.head(5)
````

````python
# Removing the column "Unnamed: 0", appears to be an extra index-volumn
# with no relevant information
chili_train = chili_train.iloc[:, 1:]
chili_test = chili_test.iloc[:, 1:]
````

### Data exploration and visualisation

````python
# Visualizing the data in a violin plot

# Looping through the dataset and displaying each feature in a violin- and
# subplot to observe the distribution and scale for each on its own.

for column in chili_train.columns:
    sns.set(rc={"figure.figsize":(8, 5)})
    
    fig , (ax_violin, ax_box) = plt.subplots(2, sharex = False, figsize=(10, 8))
    plt.gca().set(xlabel=column, ylabel='Values')
    sns.violinplot(chili_train[column], ax=ax_violin, palette="Blues")
    sns.histplot(chili_train[column], ax=ax_box, bins = 5, kde = True)
````
